```python
import os
import shutil
import random

# Define paths
source_folder = "yolo_format/obj_train_data"
image_destination_folder = "C:/Users/barbr/PycharmProjects/datasets/yolo_format/obj_train_data/images"
annotation_destination_folder = "C:/Users/barbr/PycharmProjects/datasets/yolo_format/obj_train_data/labels"

train_image_folder = os.path.join(image_destination_folder, "train")
val_image_folder = os.path.join(image_destination_folder, "val")

train_annotation_folder = os.path.join(annotation_destination_folder, "train")
val_annotation_folder = os.path.join(annotation_destination_folder, "val")

# Create destination folders if they don't exist
os.makedirs(train_image_folder, exist_ok=True)
os.makedirs(val_image_folder, exist_ok=True)
os.makedirs(train_annotation_folder, exist_ok=True)
os.makedirs(val_annotation_folder, exist_ok=True)

# Move images to train and val folders
for filename in os.listdir(source_folder):
    name, ext = os.path.splitext(filename)
    if ext.lower() in [".jpg", ".jpeg", ".png", ".bmp"]:
        if random.random() < 0.8:  # 80% train, 20% val
            destination = train_image_folder
        else:
            destination = val_image_folder
        shutil.move(os.path.join(source_folder, filename), os.path.join(destination, filename))

# Move annotations to train and val folders based on corresponding images
for filename in os.listdir(source_folder):
    name, ext = os.path.splitext(filename)
    if ext.lower() in [".xml", ".json", ".txt"]:
        corresponding_image = name + '.jpg'
        if os.path.exists(os.path.join(train_image_folder, corresponding_image)):
            destination = train_annotation_folder
        elif os.path.exists(os.path.join(val_image_folder, corresponding_image)):
            destination = val_annotation_folder
        else:
            continue
        shutil.move(os.path.join(source_folder, filename), os.path.join(destination, filename))
```

```python
import os
import shutil
import random

# Define paths
train_image_folder = "C:/Users/barbr/PycharmProjects/YOLOv8 Implementation/images/train"
val_image_folder = "C:/Users/barbr/PycharmProjects/YOLOv8 Implementation/images/val"
train_annotation_folder = "C:/Users/barbr/PycharmProjects/YOLOv8 Implementation/annotations/train"
val_annotation_folder = "C:/Users/barbr/PycharmProjects/YOLOv8 Implementation/annotations/val"

# Function to move files with specific extensions and matching annotations
def move_matched_files(src_image_folder, src_annotation_folder, dest_train_folder, dest_val_folder, extensions, split_ratio=0.8):
    print(f"Moving matched files from {src_image_folder} and {src_annotation_folder}")
    moved_count = 0

    # Collect all annotation files
    annotation_files = os.listdir(src_annotation_folder)

    for filename in os.listdir(src_image_folder):
        name, ext = os.path.splitext(filename)
        
        # Check if the corresponding annotation file exists
        annotation_filename = name + '.txt'  # Assuming annotations have the same filename as images with .txt extension
        if annotation_filename in annotation_files:
            if ext.lower() in extensions:
                if random.random() < split_ratio:
                    destination_image = dest_train_folder
                    destination_annotation = train_annotation_folder
                else:
                    destination_image = val_image_folder
                    destination_annotation = val_annotation_folder

                # Move image file
                src_image_path = os.path.join(src_image_folder, filename)
                dest_image_path = os.path.join(destination_image, filename)
                shutil.move(src_image_path, dest_image_path)

                # Move corresponding annotation file
                src_annotation_path = os.path.join(src_annotation_folder, annotation_filename)
                dest_annotation_path = os.path.join(destination_annotation, annotation_filename)
                shutil.move(src_annotation_path, dest_annotation_path)

                print(f"Moved {filename} and its annotation to {destination_image} and {destination_annotation}")
                moved_count += 1

    print(f"File moving completed. Moved {moved_count} files with matching annotations.")

# Move matched files to train and val folders
move_matched_files(train_image_folder, train_annotation_folder, train_image_folder, val_image_folder, [".jpg", ".jpeg", ".png", ".bmp"])
move_matched_files(val_image_folder, val_annotation_folder, train_image_folder, val_image_folder, [".jpg", ".jpeg", ".png", ".bmp"])

# Function to count files in a folder
def count_files(folder):
    return len([name for name in os.listdir(folder) if os.path.isfile(os.path.join(folder, name))])

# Count files in each folder after moving
train_images_count = count_files(train_image_folder)
val_images_count = count_files(val_image_folder)
train_annotations_count = count_files(train_annotation_folder)
val_annotations_count = count_files(val_annotation_folder)

# Print counts
print(f"Train Images: {train_images_count}, Train Annotations: {train_annotations_count}")
print(f"Val Images: {val_images_count}, Val Annotations: {val_annotations_count}")
```

```python
# Verify the dataset structure
def verify_paths(paths):
    for path in paths:
        if not os.path.exists(path):
            print(f"Directory does not exist: {path}")
        else:
            print(f"Directory exists: {path}")
            if not os.listdir(path):
                print(f"No files found in directory: {path}")
            else:
                print(f"Files found in directory: {path}")

dataset_dirs = [
    train_image_folder,
    val_image_folder,
    train_annotation_folder,
    val_annotation_folder
]

verify_paths(dataset_dirs)
```

```python
from ultralytics import YOLO
import yaml
import os

# Define paths for images and annotations
source_image_folder = "yolo format/obj_train_data"
train_image_folder = os.path.join(source_image_folder, "train")
val_image_folder = os.path.join(source_image_folder, "val")

# Define the dataset configuration
dataset_config = {
    'train': train_image_folder,
    'val': val_image_folder,
    'nc': 2,  # Number of classes
    'names': ['Saloon Cars', 'Motorcycles']  # Class names
}

# Save the configuration to a YAML file
ROOT_DIR = "C:/Users/barbr/PycharmProjects/YOLOv8 Implementation"
config_file_path = os.path.join(ROOT_DIR, 'data.yaml')
with open(config_file_path, 'w') as file:
    yaml.dump(dataset_config, file, default_flow_style=False)

print(f"data.yaml file created successfully at {config_file_path}.")
```

```python
# Load a model
model = YOLO("yolov8n.yaml")  # assuming "yolov8n.yaml" is your YOLO configuration file

# Full path to the YAML configuration file
data_yaml_path = config_file_path

# Use the model to train
results = model.train(data=data_yaml_path, epochs=2)  # train the model
```
